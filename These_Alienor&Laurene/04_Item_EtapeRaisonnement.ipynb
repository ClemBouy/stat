```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fanalysis.ca import CA
from matplotlib.ticker import PercentFormatter

data = pd.read_csv('data.csv')
```

### Pie charts décrivant l'étape du raisonnement clinique pour chaque item. 

Les items sont présentées par ordre décroissant. 

```python
data_gpByItem = data.groupby(['ecn_item']).agg(
    diagnostic = ('step_diag', 'sum'),
    prescription = ('step_PEC_TTT', 'sum'), 
    autre = ('step_autre', 'sum'),
    ligne_count = ('record_id', lambda x: x.count())
)

palette = sns.color_palette(n_colors=3, palette="Set2")

color_map = {
    'diagnostic' : palette[0],
    'prescription' : palette[1],
    'autre': palette[2]
}

data_gpByItem = data_gpByItem.sort_values(by="ligne_count", ascending=False)

display(data_gpByItem)

cols = data_gpByItem.columns

data_gpByItemProp = data_gpByItem.copy()
data_gpByItemProp[cols] = data_gpByItem[cols].div(data_gpByItem['ligne_count'], axis=0)

display(data_gpByItemProp)

for i in range (0,135):
    row = data_gpByItemProp.iloc[[i]]
    row = row.drop(columns='ligne_count')
    row = row.loc[:, (row != 0).any(axis=0)]

    active_cols = row.columns
    colors = [color_map[col] for col in active_cols]

    print(data_gpByItem.iloc[i])
    plt.figure(figsize=(3, 3))
    plt.pie(row.values[0], labels=row.columns, colors=colors, autopct='%1.1f%%')
    plt.title(f"Pie chart pour l'{data_gpByItem.index[i]}")
    plt.show()  
```
