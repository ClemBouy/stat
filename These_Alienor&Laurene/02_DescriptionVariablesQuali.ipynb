```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data.csv')
```

```python
categorical_columns = ['tag', 'FS', 'PE', 'RC', 'specialty', 'step_5', 'step_2', 'ecn_item', 'skill', 'ped_ambu', 'order_1', 'step_PEC_TTT', 'step_diag', 'step_autre']

for col in categorical_columns:
    display(data[col].describe())
    plt.figure(figsize=(20, 10))
    proportions = data[col].value_counts(normalize=True) * 100
    # proportions = proportions.head(15)
    print(proportions)
    sns.barplot(x=proportions.index, y=proportions.values)
    plt.title(f"Diagramme en barres pour la variable {col} (proportions)")
    plt.ylabel('Proportion (%)')
    plt.xticks(rotation=90)
    plt.show()
```
