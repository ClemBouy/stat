```python
import pandas as pd 
import numpy as np
import os
```

```python
def returnSeries(tags):
    return tags

def rename_duplicates(tag_list):
    if (isinstance(tag_list, str)): 
        return tag_list
    counts = {}
    new_list = []
    for tag in tag_list:
        if tag in counts:
            counts[tag] += 1
            new_list.append(f"{tag}#{counts[tag]}")
        else:
            counts[tag] = 1
            new_list.append(f"{tag}#1")
    return new_list

def rename_by_student(group):
    counts = {}
    new_tags = []
    for tag in group:
        if tag in counts:
            counts[tag] += 1
            new_tags.append(f"{tag}-{counts[tag]}")
        else:
            counts[tag] = 1
            new_tags.append(f"{tag}-1")
    return new_tags

def check_order(tags, tag1, tag2):
    try: 
        if (isinstance(tags, str) and (tags == tag1 or tags == tag2)):
            return tags
        tags = tags.tolist()
        
        idx_tag1 = tags.index(tag1) if tag1 in tags else float("inf")
        idx_tag2 = tags.index(tag2) if tag2 in tags else float("inf")

        if idx_tag1 < idx_tag2:
            return tag1 + " avant " + tag2
        elif idx_tag2 < idx_tag1:
            return tag2 + " avant " + tag1
        else:
            return "Aucun des deux n'a été réalisé"
        
    except Exception as e:
        return None
    
def check_value_in_array(array): 
    if 'oui' in array.values:
        return 'oui'
    return 'non'
    
def ped_ambu(x):
    if x == 'Pôle enfantinternshipPractical':
        return 'oui'
    return 'non'

def test(x): 
    print(x)

short_ids = {}

def get_short_id(uuid, col_name):
    if uuid not in short_ids[col_name]:
        short_ids[col_name][uuid] = len(short_ids[col_name]) + 1
    return short_ids[col_name][uuid]
```

```python
records = pd.read_csv('data/records_from_p22.csv')
internships = pd.read_csv('data/internships.csv')
skills_certified = pd.read_csv('data/skills_certified.csv')
consent_list = pd.read_csv('data/consent_list.csv')

internships['Ped_ambu'] = internships['tag'] + internships['type']
internships['Ped_ambu'] = internships['Ped_ambu'].apply(ped_ambu)

skills_certified = skills_certified.groupby(by='record_id').agg({'skill_code': returnSeries})

internships = internships.drop(columns='period')

internships["internship"] = internships.groupby("student_id")["tag"].transform(rename_by_student)

display(consent_list)
display(internships)
display(skills_certified)
```

```python
internships_gb = internships.groupby(by='student_id').agg({'tag' : returnSeries, 'Ped_ambu': check_value_in_array})

internships_gb["order_SN1"] = internships_gb["tag"].apply(check_order, args={"SN1", "Urgences"})
internships_gb["order_SN2"] = internships_gb["tag"].apply(check_order, args={"SN2", "Pôle femme"})

# internships_gb["tag"] = internships_gb["tag"].apply(rename_duplicates)

display(internships_gb)
```

```python
records_SN1 = records.loc[records['internship'] == 'SN2']

# data_in_rework = data.loc[data['workflow_state'] == 'in_rework']

display(records)
display(records_SN1)

records_from_refused = records_SN1.loc[records_SN1['clinical_validation_state'] == "validation_refused"]
display(records_from_refused)
```

```python
data = pd.merge(records_SN1, internships_gb, on="student_id", how="left")
data = pd.merge(data, internships, on="internship_id", how="left")
data = pd.merge(data, skills_certified, on="record_id", how="left")

data = data.drop(columns={"student_id_y", "Ped_ambu_y", "tag_y", "internship_x", "type"})
data = data.rename(columns={
    "Ped_ambu_x": "Ped_ambu",
    "tag_x": "internship_list",
    "student_id_x": "student_id",
    "internship_y": "internship",
    "skill_code": "skills_certified"
})


data = pd.merge(data, consent_list, on="student_id")

display(data)


data = data[['student_id', 'internship_id', 'record_id', 'internship', 'title', 'content', 'workflow_state', 'clinical_validation_state', 'FS', 'EP', 'internship_list', 'Ped_ambu', 'order_SN1', 'order_SN2', 'skills_certified']]


# On ne garde que les SA validés ou signés. 
data_rejected = data.loc[data['workflow_state'] == 'rejected']
print("SA refusés : " + str(len(data_rejected.index)))

data_draft = data.loc[data['workflow_state'] == 'draft']
print("SA en brouillon : " + str(len(data_draft.index)))

data_in_review = data.loc[data['workflow_state'] == 'in_review']
print("SA en demande de validation : " + str(len(data_in_review.index)))

data_in_rework = data.loc[data['workflow_state'] == 'in_rework']
print("SA en demande d'amélioration : " + str(len(data_in_rework.index)))

data = data.drop(data.loc[data['workflow_state'] == 'rejected'].index)
data = data.drop(data.loc[data['workflow_state'] == 'draft'].index)
data = data.drop(data.loc[data['workflow_state'] == 'in_review'].index)
data = data.drop(data.loc[data['workflow_state'] == 'in_rework'].index)

# Appliquer la transformation à chaque colonne
for col in ["student_id", "internship_id", "record_id"]:
    short_ids[col] = {}  # Initialiser le mapping pour cette colonne
    data[f"{col}_short"] = data[col].apply(lambda x: get_short_id(x, col))

data = data[['student_id_short', 'student_id', 'internship_id_short', 'internship_id', 'record_id_short', 'record_id', 'internship', 'title', 'content', 'workflow_state', 'clinical_validation_state', 'FS', 'EP', 'internship_list', 'Ped_ambu', 'order_SN1', 'order_SN2', 'skills_certified']]

export_command = data[['student_id_short', 'student_id', 'internship_id_short', 'internship_id', 'record_id_short', 'record_id']]
export_command.to_csv('export_command.csv')

data.to_clipboard() 
display(data)
```
